Dataset from the 2017 World Happiness Report published by the United Nations Sustainable Development Solutions Network.

This data was collected by the Gallup World Poll, using a set of surveys from more than
160 countries in over 140 languages. The surveys aim to analyze data from hundreds of thousands of individuals and
investigate the ways in which elements of people’s working lives drive their wellbeing.

The main life evaluation question asked in these surveys is the following:

“Please imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best
possible life for you and the bottom of the ladder represents the worst possible life for you. On which step of the ladder
would you say you personally feel you stand at this time?”

Each country is given a happiness score that is the sum of scores from the following 7 criteria:

    GDP per capita
    healthy years of life expectancy, social support (as measured by having someone to count on in times of trouble),
    trust (as measured by a perceived absence of corruption in government and business)
    perceived freedom to make life decisions
    generosity (as measured by recent donations)
    dystopia (benchmark country corresponding to the lowest possible values in each of the criteria)

The countries are then ranked from most happy to least happy.

Additionally, the report contains information about job satisfaction in each country.

The goal will be to investigate the relationships between these different criteria and create meaningful insights which
will be then communicate through efficient visualizations and numerical summaries.

### H. Matrices

H.1.Define a matrix whose rows correspond to countries and the columns to the regions.

```python
data_h = df.copy()
data_h = data_h.set_index('Country')      # Setting countries as index to create a new data frame with the same index

countries_region = pd.DataFrame(index=data_h.index)      # Creating Data Frame with index countries 


# Loops for 0's and 1's to identify to which region belongs each country.


W_Europe = []                   # Accumulator Pattern
for r in df['Region']:
    if r == 'Western Europe':   # if the region is equal to Western Europe, add 1 in the List W_Europe
        W_Europe.append(1)
    else:                       # else add 0
        W_Europe.append(0)

        
E_Europe = []   
for r in df['Region']:
    if r == 'Eastern Europe':
        E_Europe.append(1)
    else:
        E_Europe.append(0)

        
L_America = []
for r in df['Region']:
    if r == 'Latin America':
        L_America.append(1)
    else:
        L_America.append(0)

        
N_America = []
for r in df['Region']:
    if r == 'North America':
        N_America.append(1)
    else:
        N_America.append(0)  

Africa = []
for r in df['Region']:
    if r == 'Africa':
        Africa.append(1)
    else:
        Africa.append(0)   

        
Asia = []
for r in df['Region']:
    if r == 'Asia-Pacific':
        Asia.append(1)
    else:
        Asia.append(0)
        
# Filling DataFrame

countries_region['Western_Europe'] = W_Europe
countries_region['Eastern_Europe'] = E_Europe
countries_region['Latin_America'] = L_America
countries_region['North_America'] = N_America
countries_region['Africa'] = Africa
countries_region['Asia'] = Asia
```
